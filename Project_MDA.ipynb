In this section I used regular expression to find filing reported in a cetrrain year - there are NO reports from 2021. (If the year is chaged to so other - e.g.: 2001) 
there are files from the year visisble.

```python
import re 

pattern = r"^CONFORMED PERIOD OF REPORT:\s*2001"
```

```python
import os

matched = []

path = r"sec-edgar-filings"

for roots, dirs, files in os.walk(path):
    for file in files: 
        file_path = os.path.join(roots, file)
        f = open(file_path, "r")
        for line in f:
            if re.match(pattern, line): 
                matched.append(file_path)

print(matched)

            
        
```
